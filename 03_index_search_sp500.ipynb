# Index Search
        
        Scan an index universe (S&P 500, Nasdaq-100, Dow 30) and rank tickers using the same prediction pipeline.


```python
import analysis_blocks
```

## Block 1: Parameters


```python
index_name = "sp500"  # options: sp500, nasdaq100, dow30
limit = 500  # reduce for quicker scans
include_fundamentals = True
include_sentiment = False  # True requires API keys and adds latency
ticker_cache_dir = "cache"
ticker_cache_max_age_hours = 24
```

## Block 2: Run Index Scan


```python
result = analysis_blocks.run_index_search_workflow(
    index_name=index_name,
    limit=limit,
    include_sentiment=include_sentiment,
    include_fundamentals=include_fundamentals,
    use_ticker_cache=True,
    ticker_cache_dir=ticker_cache_dir,
    ticker_cache_max_age_hours=ticker_cache_max_age_hours,
)
```

## Block 3: Ranked Output


```python
ranked = result["prediction_summary"]
signal_only = ranked[ranked["Signal_Text"] != "HOLD"].reset_index(drop=True)

signal_only.to_csv("output/03_index_search_sp500.csv", index=False)

signal_only.head(500)
```
